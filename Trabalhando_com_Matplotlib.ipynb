## Gráfico de Linhas

```python
#Determinar o tamanho da figura
plt.figure(figsize=(9,6))

#Tipo de grafico
plt.plot(x, y)
plt.show()
```

```python
#Determinar o tamanho da figura
plt.figure(figsize=(9,6))

#Definindo a fonte padrão
fonte = {'family': 'serif',
        'color':  'darkblue',
        'weight': 'normal',
        'size': 12,
        }

#Tipo de grafico
plt.plot([1,2,3,4,5,6,7,8,9], [1,4,9,16,25,36,49,64,81])
plt.ylabel("f(x) = x² (y variavel dependente)", fontdict= fonte)
plt.xlabel("x (variavel independente)", fontdict=fonte)
plt.title("Gráfico teste", 
          loc = "center",
          fontdict = fonte)
plt.show()
```

```python
#Dois gráficos em um
x1 = np.array(x)

y1 = x1**2
y2 = x1**3

#Plotando os 2 gráficos
plt.figure(figsize=(9,6))
plt.plot(x1,y1, "o", x1,y2,"g--")

#Limitando os eixos
plt.axis([2,8,0,300])
plt.show()
```


## Plotando gráficos de Barras horizontais e vertificais


No método subplot informamos que queremos uma única figura com dois elementos ax1 para o primeiro gráfio e ax2 para o segundo, e essa figura vai ter uma largura de 10 polegadas e uma altura de 5 polegadas.

O método barh é responsável por gerar gráficos de barras horizontais. Cada gráfico tem seu próprio título e um rótulo para cada eixo.

Note que não temos mais aquele fundo branco, agora temos um fundo com grids, essa é umas das características do estilo que estamos utilizando.

```python
# Define o stilo para ggplot
plt.style.use("ggplot")
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,5))

ax1.barh(df.Sex.unique(), df.Sex.value_counts(), color = "blue")
ax2.bar(df.Cholesterol.unique(), df.Cholesterol.value_counts(), color = '#00FF00')

ax1.set(title="Gráfico de Barras Verticais", xlabel="Eixo x", ylabel=" Eixo y")
ax2.set(title="Gráfico de Barras Horizontais", xlabel="Eixo x", ylabel="Eixo y")

plt.show()
```

### Gráfico de barras com varios elementos

```python
#grandes clubes de SP e a quantidade de titulos por décadas, desde a década de 60.
Santos = [6,0,0,0,3]
Corinthians = [0,0,0,4,5]
Palmeiras = [4,2,4,3,5]
SaoPaulo = [0,1,1,1,3]
```

```python
plt.figure(figsize=(9,6))

x1 =  np.arange(len(Santos))
x2 = [x + 0.2 for x in x1]
x3 = [x + 0.2 for x in x2]
x4 = [x + 0.2 for x in x3]

# Plota as barras
plt.bar(x1, Santos, width=0.2,label = 'Santos', color = 'gray',)
plt.bar(x2, Corinthians, width=0.2, label = 'Corinthians', color = 'y')
plt.bar(x3, Palmeiras, width=0.2, label = 'Palmeiras', color = "g")
plt.bar(x4, SaoPaulo, width=0.2, label = 'São Paulo', color = 'r')

# coloca o nome dos meses como label do eixo x
decadas = ['Anos 60-69','Anos 70-79','Anos 80-89','Anos 90 - 2000','Anos 2000 - 2021']
plt.xticks([x + 0.2 for x in range(len(Santos))], decadas, rotation = 45)

# inseri uma legenda no gráfico
plt.legend()
plt.title("Titulos nacionais por década")

plt.show()
```

### Gráfico de Pizza

No nosso exemplo estamos mostrando a proporção de titulos para cada time. Fica mais fácil de visualizar quais foram os times que tiveram mais titulos nacionais

Ele é muito utilizado para mostrar a proporção de sexo (feminino / masculino) e para faixa etária (criança, jovem, adulto, idoso).

Mas se você tiver muitos dados (muitas categorias por exemplo) evite usar esse tipo de gráfico, pois a visualização fica muito confusa.

```python
plt.figure(figsize=(9,6))
Time = ["Santos", "Corinthians", "Palmeiras", "São Paulo"]
Titulos = [9,10,14,6]

plt.pie(Titulos, labels=Time, 
        colors = ["gray","y", "g","r"])
plt.show()
```

Customizando gráfico de pizza:

```python
plt.figure(figsize=(9,6))

#Destacar o primeiro elemento, o time do santos
explode = (0.2, 0, 0, 0) 
# define o formato de visualização com saída em 1.1%%, sombras e a separação entre as partes
plt.pie(Titulos, labels=Time, explode=explode, 
        colors = ["gray","y", "g","r"],
        autopct = '%1.1f%%',
        shadow = True)

plt.legend(Time, loc = 4)

plt.show()
```

Adiciona textos em gráficos

```python
# 100 valores no intervalo de 0 a 2.
x = np.linspace(0, 2, num=100)
quad = x**2

plt.figure(figsize=(9,6))
plt.plot(x, x, label="Linear", color = "r")
plt.plot(x, quad, label="Quadratica", color = "b")
plt.xlabel('x label')
plt.ylabel('y label')
plt.axis([0.5, 1.5, 0, 2])

plt.title("Gráfico de Linhas Simples")

plt.text(1, 1, "Cruzamento das Linhas", fontsize=12, horizontalalignment='right', rotation = 315)


plt.legend()



plt.show()
```
